# Détection des outliers avec la méthode interquantile

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=erpweb2['price'])
plt.xlabel("Prix",fontsize=13)
plt.xticks(fontsize=15)
plt.title("Boîte à moustache du prix",fontsize=20)
plt.show()
```

```python
def detection_outliers_interquantile(data, col):
    
    Q3 = np.quantile(data[col], 0.75)
    Q1 = np.quantile(data[col], 0.25)

    IQR = Q3 - Q1
    
    print("La valeur de Q1 pour le '%s' est de %f." % (col,Q1)) 
    
    print("La valeur de Q3 pour le '%s' est de %f." % (col,Q3))
    
    print("La valeur de l'IQR pour le '%s' est de %s." % (col, IQR))
   
    borne_inf = Q1 - 1.5 * IQR
    borne_sup = Q3 + 1.5 * IQR
    
    outliers = [x for x in data[col] if (
        (x < borne_inf) | (x > borne_sup))]
    
    notoutliers=[x for x in data[col] if (
        (x > borne_inf) & (x < borne_sup))]
    
    filtre_data = data.loc[data[col].isin(notoutliers)]
    
    print('Les outliers sont :' ,outliers)
    
    print('Dimensions du dataframe sans les outliers :',filtre_data.shape)
   
    print('Le nombre de outliers est de :',len(outliers))
```

```python
detection_outliers_interquantile(erpweb2,'price')
```

# Détection des outliers avec la méthode z-score

```python
plt.figure(figsize=(15,8))
sns.histplot((erpweb2['price']-np.mean(erpweb2['price']))/(np.std(erpweb2['price'])))
plt.title("Répartition des z-scores",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.histplot((erpweb2['price']-np.mean(erpweb2['price']))/(np.std(erpweb2['price'])))
plt.xlabel("z-score",fontsize=15)
plt.ylabel("Nombre",fontsize=15)
plt.title("Nombre de outliers",fontsize=15)
plt.axis([1.96, 7, 0, 6.5])
plt.show()
```

```python
def detection_outliers_zscore(data,col):
    seuil=1.96
    moyenne=np.mean(data[col])
    ecart_type=np.std(data[col])
    
    outliers = [x for x in data[col] if (np.abs((x-moyenne)/ecart_type)) > seuil]
    
    notoutliers=[x for x in data[col] if (np.abs((x-moyenne)/ecart_type)) < seuil]
    
    filtered_data = data.loc[data[col].isin(notoutliers)]
    
    print('Les outliers sont :' ,outliers)
    
    print('Dimensions du dataframe sans les outliers :',filtered_data.shape)
    
    print('Le nombre de outliers est de :',len(outliers))
```

```python
detection_outliers_zscore(erpweb2,'price')
```
